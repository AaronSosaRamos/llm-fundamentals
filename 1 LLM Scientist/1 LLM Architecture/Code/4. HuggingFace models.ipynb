##1. Text Classification Model

Description: A text classification model assigns predefined categories to textual data. It's widely used for sentiment analysis, spam detection, and topic labeling.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Load pre-trained model and tokenizer
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create a pipeline for sentiment analysis
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Example usage
text = "I love using Hugging Face models!"
result = classifier(text)
print(result)
```

##2. Named Entity Recognition (NER) Model

Description: NER models identify and classify entities (like names of people, organizations, locations) within text. This is essential for information extraction tasks.

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load pre-trained model and tokenizer
model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Create a pipeline for NER
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)

# Example usage
text = "Hugging Face Inc. is based in New York City."
entities = ner_pipeline(text)
print(entities)
```

##3. Text Summarization Model

Description: Summarization models condense long documents into shorter versions, capturing the main ideas. They're useful for quickly understanding large texts.

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load pre-trained model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Create a pipeline for summarization
summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)

# Example usage
text = """
Transformers have revolutionized the field of natural language processing (NLP) by introducing an architecture that excels at understanding and generating human language. The core innovation of transformers lies in their attention mechanisms, particularly the self-attention mechanism, which enables them to capture relationships between words, regardless of their position in a sentence. Unlike traditional recurrent neural networks (RNNs), transformers process input data in parallel, allowing for greater efficiency and scalability. This architecture underpins some of the most advanced large language models (LLMs) today, including GPT, BERT, and their numerous derivatives, which have set benchmarks across various NLP tasks such as translation, summarization, and conversational AI.

The adaptability of transformers makes them suitable for pretraining on massive datasets, where they learn general language representations that can later be fine-tuned for specific tasks. These models leverage billions of parameters to discern subtle nuances in text, enabling them to generate coherent and contextually relevant outputs. Their ability to handle complex linguistic tasks stems from both the depth and width of their architectures, as well as the extensive data they are trained on. This combination allows transformers to excel not only in text-based applications but also in multimodal domains, where language is integrated with images, audio, or other data types.

Despite their transformative capabilities, large language models built on transformers face challenges, including high computational costs and ethical considerations. The training and deployment of these models require significant hardware resources and energy consumption, raising concerns about environmental sustainability. Additionally, LLMs may inadvertently reinforce biases present in their training data, leading to ethical dilemmas in their deployment. Researchers are actively exploring techniques to make transformers more efficient, interpretable, and fair, ensuring their benefits can be widely and responsibly applied across industries.
"""
summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
print(summary[0]['summary_text'])
```

```python
from transformers import AutoFeatureExtractor, AutoModelForImageClassification, pipeline
from PIL import Image
import requests

# Load pre-trained model and feature extractor
model_name = "google/vit-base-patch16-224"
feature_extractor = AutoFeatureExtractor.from_pretrained(model_name)
model = AutoModelForImageClassification.from_pretrained(model_name)

# Create a pipeline for image classification
classifier = pipeline("image-classification", model=model, feature_extractor=feature_extractor)

# Example usage
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/6b/American_Beaver.jpg/640px-American_Beaver.jpg"
image = Image.open(requests.get(url, stream=True).raw)
result = classifier(image)
print(result)
```

##6. Automatic Speech Recognition (ASR) Model

Description: ASR models transcribe spoken language into text, enabling applications like voice assistants and transcription services.

```python
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torchaudio

# Load pre-trained processor and model
model_name = "facebook/wav2vec2-base-960h"
processor = Wav2Vec2Processor.from_pretrained(model_name)
model = Wav2Vec2ForCTC.from_pretrained(model_name)

# Function to resample audio to 16 kHz
def resample_audio(audio_path, target_sampling_rate=16000):
    speech_array, sampling_rate = torchaudio.load(audio_path)
    if sampling_rate != target_sampling_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=sampling_rate, new_freq=target_sampling_rate)
        speech_array = resampler(speech_array)
    return speech_array, target_sampling_rate

# Function for speech recognition
def transcribe_audio(audio_path):
    # Resample the audio to 16 kHz
    speech_array, sampling_rate = resample_audio(audio_path)

    # Process input values for the model
    inputs = processor(speech_array.squeeze(), sampling_rate=sampling_rate, return_tensors="pt", padding=True)

    # Perform inference
    logits = model(inputs.input_values).logits

    # Decode the predicted tokens
    predicted_ids = logits.argmax(dim=-1)
    transcription = processor.batch_decode(predicted_ids)[0]

    return transcription

# Example usage
audio_file_path = "/content/transformers_recording.wav"  # Replace with your audio file path
transcription = transcribe_audio(audio_file_path)
print("Transcription:", transcription)
```

```python
import torchaudio
from speechbrain.pretrained import Tacotron2, HIFIGAN
# Initialize TTS (Tacotron2) and Vocoder (HiFIGAN)
tacotron2 = Tacotron2.from_hparams(source="speechbrain/tts-tacotron2-ljspeech", savedir="tmpdir_tts")
hifi_gan = HIFIGAN.from_hparams(source="speechbrain/tts-hifigan-ljspeech", savedir="tmpdir_vocoder")
# Define the input text
text = "Hello, this is a text-to-speech test using FastSpeech2 and HiFIGAN."

# Generate the mel-spectrogram from text
mel_output, mel_length, alignment = tacotron2.encode_text(text)

# Synthesize speech from the mel-spectrogram
waveform = hifi_gan.decode_batch(mel_output)

# Save the output waveform
output_path = "output.wav"
torchaudio.save(output_path, waveform.squeeze(1), 22050)
```

##8. Image Captioning Model

Description: Image captioning models generate descriptive text based on the content of an image. These models combine computer vision (to extract image features) and natural language generation (to produce captions). They are commonly used in accessibility tools, digital asset management, and AI applications requiring scene understanding.

image.png

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import requests

# Load the processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")

# Load and process the image
image_url = "https://cdn0.expertoanimal.com/es/posts/5/8/1/cuidados_de_una_capibara_20185_orig.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)
inputs = processor(images=image, return_tensors="pt")

# Generate caption
outputs = model.generate(**inputs)
caption = processor.decode(outputs[0], skip_special_tokens=True)
print(caption)
```

```python
from transformers import AutoProcessor, AutoModelForCausalLM
from PIL import Image
import requests
import torch

# Load the processor and model
processor = AutoProcessor.from_pretrained("microsoft/git-large-coco")
model = AutoModelForCausalLM.from_pretrained("microsoft/git-large-coco")

# Load and process the image
image_url = "https://cdn0.expertoanimal.com/es/posts/5/8/1/cuidados_de_una_capibara_20185_orig.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)
pixel_values = processor(images=image, return_tensors="pt").pixel_values

# Generate caption
generated_ids = model.generate(pixel_values=pixel_values, max_length=50)
caption = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(caption)
```

```python
from transformers import VisionEncoderDecoderModel, ViTImageProcessor, AutoTokenizer
from PIL import Image
import requests
import torch

# Load the model, processor, and tokenizer
model = VisionEncoderDecoderModel.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
feature_extractor = ViTImageProcessor.from_pretrained("nlpconnect/vit-gpt2-image-captioning")
tokenizer = AutoTokenizer.from_pretrained("nlpconnect/vit-gpt2-image-captioning")

# Load and process the image
image_url = "https://cdn0.expertoanimal.com/es/posts/5/8/1/cuidados_de_una_capibara_20185_orig.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)
pixel_values = feature_extractor(images=image, return_tensors="pt").pixel_values

# Generate caption
generated_ids = model.generate(pixel_values, max_length=50)
caption = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(caption)
```
